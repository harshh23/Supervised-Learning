# **Gradient Descent**

**WAP in python to plot a function**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 5*np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot of y = sin(x)')
plt.grid(True)
plt.show()
```

```python
#plot a function
def plot_function(func):
    x_values = np.linspace(-10, 10, 1000)
    y_values=[func(x) for x in x_values]
    #y=x**2
    print("Values of x: ", x_values[0:999:100])
    print("Values of y: ", y_values[0:999:100])
    plt.plot(x_values, y_values)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Function Plot')
    plt.grid(True)
    plt.show()

plot_function(lambda x:x**2)
```

```python
def plot_derivative(func):
    x_values = np.linspace(-10, 10, 1000)
    delta_x=0.001
    y_values=[(func(x+delta_x)-func(x))/delta_x for x in x_values]
    plt.plot(x_values, y_values)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Derivative Plot')
    plt.grid(True)
    plt.show()

plot_derivative(lambda x:x**2)
```

```python
#gradient descent
def gradient_descent(func, initial_guess, learning_rate, num_iterations):
    x = initial_guess
    x_values = [x]
    for _ in range(num_iterations):
        gradient = (func(x + learning_rate) - func(x)) / learning_rate
        x -= learning_rate * gradient
        x_values.append(x)
    return x_values

gradient_descent(lambda x:x**2, 5, 0.1, 10)
```

```python
#plot gradient descent
def plot_gradient_descent(func, initial_guess, learning_rate, num_iterations):
    x_values = gradient_descent(func, initial_guess, learning_rate, num_iterations)
    y_values = [func(x) for x in x_values]
    plt.plot(x_values, y_values, marker='o')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gradient Descent')
    plt.grid(True)
    plt.show()

plot_gradient_descent(lambda x:x**2, 5, 0.1, 10)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, lambdify

def function():
    func=input("Enter a function of x (e.g., x**2 + 3*x + 2): ")
    return func

def calculate_derivative(func_str):
    x=symbols('x')
    func_expr = eval(func_str)
    derivative = diff(func_expr, x)
    return derivative

def plot_function_and_derivative(func_str, derivative, x_values):
    x = symbols('x')
    func = lambdify(x, eval(func_str), 'numpy')
    derivative_func = lambdify(x, derivative, 'numpy')
    y_values = func(x_values)
    y_derivative_values = derivative_func(x_values)
    plt.plot(x_values, y_values, label='Function: ' + func_str)
    plt.plot(x_values, y_derivative_values, label='Derivative: ' + str(derivative))
    plt.title('Function and its Derivative')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()

func_str = function()
derivative = calculate_derivative(func_str)
x_values = np.linspace(-10, 10, 400)
plot_function_and_derivative(func_str, derivative, x_values)
```

```python
def f(x):
    return x**2
def gradient(x):
    return 2 * x
def gradient_descent(learning_rate, num_iterations, start_point):
    x = start_point
    x_values = [x]
    for _ in range(num_iterations):
        grad = gradient(x)
        x = x - learning_rate * grad
        x_values.append(x)
    return x_values

learning_rate = 0.1
num_iterations = 25
start_point = 10
x_values = gradient_descent(learning_rate, num_iterations, start_point)
```

```python
x_range = np.linspace(-10, 10, 400)
y_range = f(x_range)

plt.plot(x_range, y_range, label="f(x) = x^2")
plt.scatter(x_values, [f(x) for x in x_values], color='red', label="Gradient Descent Steps")
plt.title("Gradient Descent on f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
