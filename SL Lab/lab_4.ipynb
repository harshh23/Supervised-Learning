Dataset used: Salary_Data.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/SL/Salary_Data.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
y=df['Salary']
x=df.drop('Salary', axis=1)
y.shape,x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=1)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_test_predict=model.predict(x_test)
```

```python
y_test_predict
```

```python
model.score(x_test,y_test)
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
plt.scatter(y_test_predict,y_test)
plt.legend()
plt.show()
```

```python
import seaborn as sns
imp=pd.DataFrame(list(zip(x_test.columns,np.abs(model.coef_))),columns=['feature','coeff'])
sns.barplot(x='feature',y='coeff',data=imp)
plt.xticks(rotation=90)
```
